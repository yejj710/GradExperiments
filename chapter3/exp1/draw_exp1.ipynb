```python
import matplotlib.pyplot as plt
import numpy as np

categories = ['plain', 'LDP', 'PHE', 'Semi2k', 'TEE']

values = [30.902023553848267, 34.65807223320007, 1983.3110427856445, 974.8450841903687, 51.8430219264593]
values = np.array(values) / 3

plt.bar(categories, values, color='black')

plt.xlabel('Agg Type')
plt.ylabel('Cost-Time per epoch')
plt.title('Bar Chart Example')
plt.legend(['Values'])
plt.show()
```
